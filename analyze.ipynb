```julia
using DataFrames, Plots, Statistics, DSP, RollingFunctions, FINUFFT, FFTW, StatsPlots
import CSV
```

```julia
all_values = vcat(let last_offset = 0
    map(filter(isfile, ["sample_values.csv" "values.csv" "values1.csv" "values3.csv" "values4.csv"])) do fn
        values = CSV.read(fn, DataFrame)
        rename!(values, [:t, :i])

        values.t .-= values.t[1]
        values.t .+= last_offset
        last_offset = values.t[length(values.t)] + median(values.t[2:length(values.t)] .- values.t[1:length(values.t) - 1])
        values
    end
end...);
```

```julia
plot(all_values.t, all_values.i)
```

```julia
plot(all_values.t, all_values.i, xlim=(3250000,3250000 + 200 * 641), marker=2)
```

```julia
histogram(all_values.i, yaxis=:log10)
```
