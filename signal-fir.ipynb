```julia
using DataFrames, Plots, Statistics, DSP, RollingFunctions
import CSV
```

```julia
values = CSV.read("values.csv", DataFrame);
rename!(values, [:t, :i]);
```

```julia
r=700:750; plot(values.t[r], [values.i[r]], marker = 2)
```

```julia
plot(values.t[r], [values.i[r], (runmax(values.i[r], 2) .+ runmin(values.i[r], 2)) .* 0.5], marker=2, label = ["x" "mid"])
```

```julia
p = Periodograms.periodogram(values.i)
plot(p.freq, p.power, yscale=:log10)
```

```julia
1/(median(values.t[2:size(values)[1]] .- values.t[1:(size(values)[1] - 1)]) * 1e-6)
```

```julia
histogram((values.t[2:size(values)[1]] .- values.t[1:(size(values)[1] - 1)]), normed = true)
```

```julia
responsetype = Lowpass(200; fs=500)
designmethod = FIRWindow(hanning(64))
plot(values.t, filt(digitalfilter(responsetype, designmethod), values.i))
```
